```python
import matplotlib.pyplot as plt
import numpy as np
from keras.utils import np_utils
```

```python
xtrain = np.load("X_train (3).npy")
ytrain_ = np.load("Y_train (3).npy")
xtest = np.load("X_test.npy")
```

```python
ytrain = np_utils.to_categorical(ytrain_)
```

```python
ytrain = ytrain[:200]
```

```python
xtrain = xtrain[:200]
```

```python
print(xtrain.shape,ytrain.shape)
xtrain = np.array(xtrain)
ytrain = np.array(ytrain)
#xtest = np.array(xtest)
```

```python
from keras.applications.resnet50 import ResNet50
from keras.models import Model
from keras.layers import *
```

```python
from keras.optimizers import Adam
```

```python
model = ResNet50(include_top = False , weights = "imagenet",input_shape = (227,227,3))
```

```python
model.summary()
```

```python
avg1 = GlobalAveragePooling2D()(model.output)
fc1 = Dense(256,activation = "relu")(avg1)
d1 = Dropout(0.5)(fc1)
fc2 = Dense(10,activation = "softmax")(d1)

model_new  = Model(input = model.input,outputs = fc2)
model_new.summary()
```

```python
adam = Adam(lr= 0.0003)
model_new.compile(loss = "categorical_crossentropy",optimizer = adam,metrics = ['accuracy'])
```

```python
for ix in range(len(model_new.layers)):
    print(ix,model_new.layers[ix])
```

```python
for ix in range(169):
    model_new.layers[ix].trainable = False
```

```python
model_new.compile(loss = "categorical_crossentropy",optimizer = adam,metrics = ['accuracy'])
```

```python
model_new.summary()
```

```python
model_new.fit(xtrain,ytrain,shuffle = False ,epochs = 10,batch_size = 64,validation_split =0.20)
```

```python
y_ = model_new.predict(xtest)
```

```python
pred = []
for i in y_:
    pred.append(np.argmax(i))
```
